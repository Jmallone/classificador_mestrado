```python
import numpy as np
import glob as glob
import sys

def calcular_estatisticas(resultados):
    return np.mean(resultados), np.std(resultados), np.min(resultados), np.max(resultados)

def imprimir_estatisticas(resultados):
    media, desvio, mini, maxi = calcular_estatisticas(resultados)
    print("Resultados: %.2f +- %.2f, min: %.2f, max: %.2f" % (media, desvio, mini, maxi))
    
# Redirecionar a sa√≠da para um arquivo
output_file = open('resultados.txt', 'w')

patchs = glob.glob('resultados/*')
patchs.sort()

def calcular_estatisticas(resultados):
    return np.mean(resultados), np.std(resultados), np.min(resultados), np.max(resultados)

def imprimir_estatisticas(resultados):
    media, desvio, mini, maxi = calcular_estatisticas(resultados)
    print("Resultados: %.2f +- %.2f, min: %.2f, max: %.2f" % (media, desvio, mini, maxi))
    return "Resultados: %.2f +- %.2f, min: %.2f, max: %.2f\n" % (media, desvio, mini, maxi)

for p in patchs:
    try:
        titulo = p.split('resultados/')[-1].replace('_', ' ').upper()
        print('------------------- '+titulo.rjust(10), end=' ---------------')
        print()
        output_file.write('------------------- '+titulo.rjust(10) + ' ---------------\n')
        resultados = {
            'knn': [],
            'rf': [],
            'svm': [],
            'mlp': [],
            'svm_pca=16': [],
            'knn_pca=16': [],
            'rf_pca=16': [],
            'mlp_pca=16': [],
            'svm_pca=32': [],
            'knn_pca=32': [],
            'rf_pca=32': [],
            'mlp_pca=32': [],
            'svm_pca=64': [],
            'knn_pca=64': [],
            'rf_pca=64': [],
            'mlp_pca=64': [],
        }

        for i in range(1,11):
            tmp = np.load(p+'/test'+str(i)+'_result.npy', allow_pickle=True).item()
            for k in tmp.keys():
                resultados[k].append(tmp[k])

        resultados = {k: v for k, v in resultados.items() if v}

        ''' Mostrar os Resultados'''
        for teste in sorted(resultados):
            print(teste.rjust(10), end=' - ')
            res = imprimir_estatisticas(resultados[teste])
            output_file.write(teste.rjust(10) + ' - ' + res)
        print()
        output_file.write('\n')
    except:
        continue

# Fechar o arquivo
output_file.close()
```
