```python
import os
import glob

# Caminho para o arquivo de entrada e saída
# input_file = 'VITSMALLFMD_1X1.txt'
# output_file= 'VITSMALLFMD_1X1_2.txt'

# input_file = 'FMD_obif2.txt'
# output_file= 'OBIF2FMD_1X1.txt'

patch_geral = '2'
dataset = 'FMD'

def converter(arquivo):
    print(f"Processando o arquivo [{arquivo}]")
    outro_nome = arquivo.replace('_1x1', '').replace('_2x2', '').split("_")[1].replace('.txt', '').upper()
    input_file = arquivo
    output_file= f'{outro_nome}{dataset}_{patch_geral}X{patch_geral}.txt'

    def load_labels(file_path):
        labels_dict = {}
        with open(file_path, 'r') as file:
            for line in file:
                label, nome = line.strip().split(' ')
                labels_dict[nome] = label
        return labels_dict

    def get_label(image_path, labels_dict):
        nome_arquivo = os.path.basename(image_path)
        return labels_dict.get(nome_arquivo, None)

    # Função para processar o arquivo .txt
    def process_file(input_file, output_file, labels_dict):
        with open(input_file, 'r') as infile, open('../'+output_file, 'w') as outfile:
            for line in infile:
                # Separar os componentes da linha
                parts = line.strip().split()
                # *features, nome_classe, image_path = parts
                '''
                    USEI para o DTD
                '''
#                 *features, nome_classe, image_path, part_1, part_2 = parts

                
                # *features, nome_classe, image_path, image_name, part_1, part_2 = parts
                *features, nome_classe, image_path, part_1 = parts
                # print(features, nome_classe, image_path, part_1)
                # print(features, nome_classe, image_path)
#                 print(outro_nome)
                # input("\n>>>>")

                # Obter o novo label numérico
                
                image_path = image_path.replace('_0_0.jpg', '.jpg').replace('_1_0.jpg', '.jpg').replace('_1_1.jpg', '.jpg').replace('_0_1.jpg', '.jpg')
                new_label = get_label(image_path, labels_dict)
                
#                 print(f"new_label: {new_label}")
#                 input("bbbb")

                # Se não encontrar um label, usar um valor padrão ou pular a linha
                if new_label is None:
                    print('-------------')
                    print(f"Warning: No label found for {image_path}")
                    print(arquivo)
                    print('-------------')
                    break
                    # continue

                # Montar a nova linha
                new_line = ' '.join(features) + f' {new_label} {nome_classe}/{image_path}\n'

                # Escrever a nova linha no arquivo de saída
                outfile.write(new_line)

    # Carregar os labels em um dicionário
    labels_dict = load_labels('LABELS_processed.txt')

    # Processar o arquivo
    process_file(input_file, output_file, labels_dict)

arquivos = glob.glob("*.txt")
# arquivos = [a.replace(".txt", '').replace("FMD_2x2_", '').replace('_', '').upper() for a in arquivos]


for arq in arquivos:
#     print(arq)
    if arq == "LABELS_processed.txt":
        continue
    converter(arq)
```
